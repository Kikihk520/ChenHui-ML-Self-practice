```python
import pandas as pd
score = pd.read_csv('score.csv')
score
```

```python
x = score.drop(columns=['scores'])
x
```

```python
y = score['scores']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x,y)
prediction = model.predict([[1000]])
prediction
```

```python
import joblib
joblib.dump(model,'score.joblib')
```
